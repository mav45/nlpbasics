```python
import spacy
nlp = spacy.load('en_core_web_sm')
from spacy import displacy
```

```python
my_text = nlp(u"The work order included the inclusion of electric bikes and carbon fishing poles from the "+ 
              "main office in Denver. REI and LL Bean were very excited because they ordered over 50 items")
```

```python
for np in my_text.noun_chunks:
    print(np.text)
    
```

```python
#visualize the text components though syntaxtic analysis and ne blotting 
```

```python
displacy.render(my_text, style='dep', jupyter=True)
```

```python
displacy.render(my_text, style='ent', jupyter=True)
```
